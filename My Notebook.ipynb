# Statistical Data Visualization using Plotly

In this project we will explore plotly express to visually statistical plots such as box plots, histograms, heat maps, density maps, contour plots and violent plots.

Some important background about plotly package.

- The plotly Python package empowers anyone to create, manipulate and render graphical figures.
- The figures are represented by data structures referred to as figures. 
- The rendering process uses the Plotly.js JavaScript library under the hood but you never need to use Java directly. 
- Figures can be represented in Python either as dictionaries or as instances of the plotly.graph_objects

__Note:__ 
1. Plotly Express is the recommended entry-point into the plotly package 
2. PLotly Express is the high-level plotly.express module that consists of Python functions which return fully-populated plotly.graph_objects.Figure objects. 
3. plotly.express module contains functions that can create interactive figures using a very few lines of code
4. Plotly Express is refered to as px. 
5. Plotly Express is a built-in part of the plotly library 
6. Plotly Express function uses graph objects internally and returns a plotly.graph_objects.Figure instance.
7. Check out the documentation here: https://plotly.com/python/plotly-express/ 

### TASK #1: PLOT INTERACTIVE BOX PLOT USING PLOTLY EXPRESS

```python
# A box plot is a statistical representation of numerical data through their quartiles. 
# The ends of the box represent the lower and upper quartiles, while the median (second quartile) is marked by a line inside the box. 

import plotly.express as px
import pandas as pd
import numpy as np
```

```python
# Import Cancer data drom the Sklearn library
cancer_df = pd.read_csv('cancer.csv')
```

```python
# Check out the head of the dataframe
cancer_df.head()
```

What we got here is the data sets that we're going to be working with. This is a real world data set called cancer data set. Here we can see that we have different features. These are from different breast cancer cells.

Here we have the mean radius of these cells the mean texture, mean perimeter, mean area, smoothness, and so on...

And what we can see at the last colum is the target. The target is simply a binary column indicating either zero or one which either the patient has cancer or not.
So, depending basically on the values for each of the features here,
we can tell if that patient target is classified as zero
or 1. Zero and one indicate that the tumor
is either malignant or benign meaning if the tumor is growing or not.

```python
# Check out the tail of the dataframe
cancer_df.tail()
```

We can see that the last row ended at 568. This would mean we have 569 rows or patient in our dataset.

Let's plot the boxplot.

```python
# Plot the boxplot
fig = px.box(cancer_df, x='target', y='mean area')
fig.show()
```

In the above plot we can see the distribution of our data for class "0" and class "1". If we hover on top of our plots, 
we should be able to see the median value, first quartile, Q3, Q4 and etc...

Any point beyond the lines in our plot are called outliers meaning they are just beyond the normal range of our data.

```python
# adding visibility of the datapoints in our plot above

fig = px.box(cancer_df, x='target', y='mean area', points='all')
fig.show()
```

- **Plot the boxplot for Mean Perimeter, use points = "all"** 

```python
fig = px.box(cancer_df, x='target', y='mean perimeter', points='all')
fig.show()
```

We managed to plot the box plot for "mean perimeter". For the class zero, the median values stands around 1,14, the maximum value stands around 188 and the minimum perimeter stands
at 71.

For class one we should be able to see the median stands at 78, the maximum is at 114, and minimum stands at 43.

### TASK #2: PLOT INTERACTIVE HISTOGRAMS

```python
# A histogram is representation of the distribution of numerical data, where the data are binned and the count for each bin is represented. 
# plot mean radius
fig = px.histogram(cancer_df, x = 'mean radius', nbins=60)
fig.show()
```

what we have here is a plot of of the distribution of our data for mean radius. This looks more of a normal distribution. Alot of our samples ranges between 12 to 12.49. Some samples ranges also between 11.5 to 11.99.

The y-axis here shows the count of how many samples fall within each of these bins. We can also observe here that we have some samples that range from 28-2849, which is big, but very few (below 10). We can also see the samples with the smallest size of mean radius here.

```python
# let's create a plot of 2 histograms in one plot. Let's plot for target class zero and one.
fig = px.histogram(cancer_df, x='mean radius', color='target', nbins=60)
fig.show()
```

These divisions in color are still the same exact representation of plot above. Except, we determined here which has benign and malignant by specifying the "target" column in the color argument. 

Wecan see here the big difference between the two. The mean radius for the class zero tends to be generally larger/bigger compare to class one.

- **Plot the histogram for the mean permieter for the entire dataset**

```python
fig = px.histogram(cancer_df, x='mean perimeter', nbins=60)
fig.show()
```

We can see the overall distribution of mean perimeter. Looks like normal distribution. The data mostly lies on from 70 to 90 in the x-axis. Then, the count reduce significantly when going to 180.

- **Plot the histogram for the mean perimeter for each of the class independantly**

```python
fig = px.histogram(cancer_df, x='mean perimeter', color='target', nbins=60)
fig.show()
```

Similar with our observation on the mean radius, the mean perimeter for target class 0 tend to be larger or higher compare to class 1.

### TASK #3: PLOT INTERACTIVE HISTOGRAMS WITH MARGINAL PLOTS

```python
# With the marginal keyword, a subplot is drawn alongside the histogram, visualizing the distribution. 
# For example, the plotly.express function px.histogram can add a subplot with a different statistical representation than the histogram, given by the parameter marginal. Plotly Express is the easy-to-use, high-level interface to Plotly, 
fig = px.histogram(cancer_df, x = 'mean radius', color = 'target', marginal='rug', nbins=60)
fig.show()
```

What we see here is a histogram with marginal plot. The marginal plot here is an additional representation of distribution of data. If we zoom in, we will be able to seeevery single data in our data set and we can infer better how the data can be skewed on a certain area.

```python
# let's explore another type of marginal plot. Let's use boxplot.
fig = px.histogram(cancer_df, x = 'mean radius', color = 'target', marginal='box', nbins=60)
fig.show()
```

Here we can see more than one plot in our figure. We have the histogram and a box plot on top of it. If we hover these plots, we should be able to get a statistical summary of the data in mean __radius column__.

- **Plot the histogram for the mean perimeter using 40 bins and explore a new marginal plot**

```python
fig = px.histogram(cancer_df, x = 'mean perimeter', color = 'target', marginal='violin', nbins=40)
fig.show()
```

Violin plot shows the distribution of data in symmetric.

### TASK #4: PLOT INTERACTIVE DENSITY MAP

A 2D histogram, also known as a density heatmap, is the 2-dimensional generalization of a histogram which resembles a heatmap but is computed by grouping a set of points specified by their x and y coordinates into bins, and applying an aggregation function such as count or sum (if z is provided) to compute the color of the tile representing the bin. 

This kind of visualization (and the related 2D histogram contour, or density contour) is often used to manage over-plotting, or situations where showing large data sets as scatter plots would result in points overlapping each other and hiding patterns.

```python
fig = px.density_heatmap(cancer_df, x = 'mean radius', y = 'mean texture')
fig.show()
```

The mean radius here in x-axis is a histogram. The color here represents the number of intersection of count of samples between mean radius and mean texture. If we hover also on each bin, we will find the location of the bin with their range. For example, in the mean radius of 12-13.99 and mean texture of 16-17.99, we can see that the bin with the most yellow has a 39 counts of samples.

### TASK #7: PLOT INTERACTIVE 2D HISTOGRAM CONTOUR PLOT

A 2D histogram contour plot, also known as a density contour plot, is a 2-dimensional generalization of a histogram which resembles a contour plot but is computed by grouping a set of points specified by their x and y coordinates into bins, and applying an aggregation function such as count or sum (if z is provided) to compute the value to be used to compute contours. 

This kind of visualization (and the related 2D histogram, or density heatmap) is often used to manage over-plotting, or situations where showing large data sets as scatter plots would result in points overlapping each other and hiding patterns.


```python
fig = px.density_contour(cancer_df, x='mean radius', y = 'mean texture')
fig.show()
```

Here we have a contour plot that represents some sort of aggregation of values between the mean radius and the mean texture.

If we zoom in to the center part:
  

the mean radius is around 13 and the mean texture is approximately around 17. Also, the count of samples that belong within that range is 39.

```python
# Marginal plots can be added to visualize the 1-dimensional distributions of the two variables. 
# Here we use a marginal histogram. Other allowable values are violin, box and rug.
fig = px.density_contour(cancer_df, x='mean radius', y = 'mean texture', marginal_x='histogram', marginal_y='histogram')
fig.show()
```

We were able to plot the same contour plot before but on top of that we were able to plot the distribution, the histogram for our mean radius and hstogram for the mean texture.

Let's add the color attribute.

```python
fig = px.density_contour(cancer_df, x='mean radius', y='mean texture', color='target')
fig.show()
```

Here we were able to plot each classes' countour both independently. Again, we can hopven on top of it or select the specific class we would want to focus.

```python
# Plotly Express density contours can be continuously-colored and labeled:

fig = px.density_contour(cancer_df, x='mean radius', y= 'mean texture')
fig.update_traces(contours_coloring='fill', contours_showlabels=True)
```

Here we were able to plot the same plot before but we added color coding by using contour coloring. We also show here the label by using contours showlabels.
